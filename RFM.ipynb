```python
import pandas as pd
import datetime as dt
import numpy as np

# importing the data
df = pd.read_csv('bills.csv')
df.head()
```

```python
df['to_user_date'] = df['to_user_date_ngay_giao_hang'].str.split('T',expand=True)[0]
df['create_info_date'] = df['create_info_created_ngay_dat_hang'].str.split('T',expand=True)[0]
```

```python
df['create_info_date'].max()
```

```python
df.head()
```

```python
from datetime import datetime as dt
df_recency = df.groupby(by='customer_id',
                        as_index=False)['create_info_date'].max()
df_recency.columns = ['customer_id', 'LastPurchaseDate']
df_recency['LastPurchaseDate'] = df_recency['LastPurchaseDate'].apply(lambda x: dt.strptime(x, '%Y-%m-%d'))
recent_date = df_recency['LastPurchaseDate'].max()
df_recency['Recency'] = df_recency['LastPurchaseDate'].apply(
    lambda x: (recent_date - x).days)
df_recency.head()
```

```python
frequency_df = df.drop_duplicates().groupby(
    by=['customer_id'], as_index=False)['create_info_date'].count()
frequency_df.columns = ['customer_id', 'Frequency']
frequency_df.head()
```

```python
# monetary_df = df.groupby(by='customer_id', as_index=False)['summary_grand_total'].sum()
# monetary_df.columns = ['customer_id', 'Monetary']
# monetary_df.head()
```

```python
rfm_df = df_recency.merge(frequency_df, on='customer_id')
rfm_df.head()
```

```python
rfm_df.info()
```

```python
rfm_df = rfm_df.loc[(rfm_df['LastPurchaseDate'] >= '2022-10-23')]
```

```python
rfm_df['R_rank'] = rfm_df['Recency'].rank(ascending=False)
rfm_df['F_rank'] = rfm_df['Frequency'].rank(ascending=True)
# rfm_df['M_rank'] = rfm_df['Monetary'].rank(ascending=True)

# normalizing the rank of the customers
# Assume df là DataFrame của bạn và 'R_rank', 'F_rank', 'M_rank' là các cột rank cần chuẩn hóa
rfm_df['R_rank'] = round(1 + 4 * (rfm_df['R_rank'] - rfm_df['R_rank'].min()) / (rfm_df['R_rank'].max() - rfm_df['R_rank'].min())).astype(int).astype(str)
rfm_df['F_rank'] = round(1 + 4 * (rfm_df['F_rank'] - rfm_df['F_rank'].min()) / (rfm_df['F_rank'].max() - rfm_df['F_rank'].min())).astype(int).astype(str)
# rfm_df['M_rank'] = round(1 + 4 * (rfm_df['M_rank'] - rfm_df['M_rank'].min()) / (rfm_df['M_rank'].max() - rfm_df['M_rank'].min()))

rfm_df.head()
```

```python
rfm_df['RF_Score'] = rfm_df['R_rank'] + rfm_df['F_rank']
```
